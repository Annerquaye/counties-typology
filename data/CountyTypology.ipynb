```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
pd.set_option('display.max_columns', 500)
```

```python
df = gpd.read_file("countyTypologyCodes.geojson")
```

```python
df.head(10)
```

```python
statesData = pd.pivot_table(df, index='STATE_NAME', values='Economic Types Type_2015_Update non-overlapping', columns='Economic_Type_Label', aggfunc="count").reset_index()
statesData.head(5)
```

```python
statesData['max'] = statesData[['Farming', 'Federal/State Government', 'Maufacturing', 'Mining', 'Nonspecialized', 'Recreation']].idxmax(axis=1)
statesData['max']
```

```python
statesData.rename(columns={'Maufacturing':'Manufacturing'}, inplace=True)
```

```python
dfState = gpd.read_file("statesElections.geojson")
```

```python
dfState
```

```python
statesData.rename(columns={'STATE_NAME':'State'},inplace = True)
```

```python
dfState
```

```python
mergedDF = statesData.merge(dfState, on='State')
```

```python
mergedDF.columns
```

```python
mergedDF = mergedDF[['State', 'Farming', 'Federal/State Government', 'Manufacturing',
       'Mining', 'Nonspecialized', 'Recreation', 'max', 'geometry']]
```

```python
mergedDF
```

```python
gdf = gpd.GeoDataFrame(mergedDF, crs="EPSG:4326", geometry=mergedDF.geometry)
```

```python
gdf.to_file('stateTypology.geojson', driver='GeoJSON')
```
